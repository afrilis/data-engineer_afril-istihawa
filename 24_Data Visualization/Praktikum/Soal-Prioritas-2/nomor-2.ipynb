```python
import pandas as pd

# Membaca data dari file CSV
df = pd.read_csv('survey.csv')

# Menghapus data duplikat berdasarkan kolom 'name'
df.drop_duplicates(subset='name', keep='first', inplace=True)

# Menghapus data dengan atribut yang tidak lengkap
df.dropna(subset=['age', 'favorite_front_end', 'favorite_back_end', 'salary_per_year_in_USD', 'company_type'], inplace=True)

# Menyimpan dataframe yang sudah diperbaharui ke dalam file CSV
df.to_csv('survey_updated.csv', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Membaca data dari file CSV
df = pd.read_csv('survey_updated.csv')

# Persentase responden berdasarkan age
plt.figure(figsize=(10, 6))
df['age'].value_counts(normalize=True).sort_index().plot(kind='bar', color='skyblue')
plt.title('Persentase Responden Berdasarkan Usia')
plt.xlabel('Usia')
plt.ylabel('Persentase')
plt.grid(axis='y')
plt.show()
```

```python
# Bahasa pemrograman / framework yang paling banyak digunakan dalam favorite_front_end
plt.figure(figsize=(10, 6))
df['favorite_front_end'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Bahasa Pemrograman / Framework yang Paling Banyak Digunakan dalam Favorite Front End')
plt.xlabel('Bahasa Pemrograman / Framework')
plt.ylabel('Jumlah Responden')
plt.grid(axis='y')
plt.show()
```

```python
# Bahasa pemrograman / framework yang paling banyak digunakan dalam favorite_back_end
plt.figure(figsize=(10, 6))
df['favorite_back_end'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Bahasa Pemrograman / Framework yang Paling Banyak Digunakan dalam Favorite Back End')
plt.xlabel('Bahasa Pemrograman / Framework')
plt.ylabel('Jumlah Responden')
plt.grid(axis='y')
plt.show()
```

```python
# Persentase bahasa pemrograman / framework yang dipilih dalam favorite_front_end
plt.figure(figsize=(10, 6))
df['favorite_front_end'].value_counts(normalize=True).plot(kind='bar', color='skyblue')
plt.title('Persentase Bahasa Pemrograman / Framework dalam Favorite Front End')
plt.xlabel('Bahasa Pemrograman / Framework')
plt.ylabel('Persentase')
plt.grid(axis='y')
plt.show()
```

```python
import pandas as pd

# Membaca data dari file CSV
df = pd.read_csv('survey_updated.csv')

# Mengubah rentang gaji menjadi nilai tengah dari setiap rentang
df['salary_per_year_in_USD'] = df['salary_per_year_in_USD'].apply(lambda x: sum(map(int, x.split(' - ')))/2 if 'more than' not in x else int(x.split(' ')[2]))
```

```python
# Rentangan gaji per tahun yang paling banyak berdasarkan salary_per_year_in_USD
plt.figure(figsize=(10, 6))
df['salary_per_year_in_USD'].plot(kind='hist', bins=10, color='skyblue')
plt.title('Rentangan Gaji per Tahun yang Paling Banyak')
plt.xlabel('Gaji per Tahun (USD)')
plt.ylabel('Jumlah Responden')
plt.grid(axis='y')
plt.show()
```
